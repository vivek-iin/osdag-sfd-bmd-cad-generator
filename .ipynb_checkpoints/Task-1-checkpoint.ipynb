```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("SFS_Screening_SFDBMD.xlsx.csv")

# Extract relevant columns
positions = df["Distance (m)"]
shear_force = df["SF (kN)"]
bending_moment = df["BM (kN-m)"]

# Create the plots
plt.figure(figsize=(12, 5))

# Shear Force Diagram (SFD)
plt.subplot(1, 2, 1)
plt.plot(positions, shear_force, marker='o', linestyle='-', color='b', label="Shear Force")
plt.axhline(0, color='black', linewidth=1)
plt.xlabel("Position along Beam (m)")
plt.ylabel("Shear Force (kN)")
plt.title("Shear Force Diagram (SFD)")
plt.legend()
plt.grid()

# Bending Moment Diagram (BMD)
plt.subplot(1, 2, 2)
plt.plot(positions, bending_moment, marker='o', linestyle='-', color='r', label="Bending Moment")
plt.axhline(0, color='black', linewidth=1)
plt.xlabel("Position along Beam (m)")
plt.ylabel("Bending Moment (kN-m)")
plt.title("Bending Moment Diagram (BMD)")
plt.legend()
plt.grid()

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Excel data (update file path)
file_path = "SFS_Screening_SFDBMD.xlsx.csv"
df = pd.read_excel(file_path)

# Extract data
x = df["Position"]
shear_force = df["Shear Force"]
bending_moment = df["Bending Moment"]

# Create figure and subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot Shear Force Diagram
axes[0].plot(x, shear_force, color='b', marker='o', linestyle='-')
axes[0].axhline(0, color='black', linewidth=0.8)
axes[0].set_title("Shear Force Diagram (SFD)")
axes[0].set_xlabel("Beam Length (m)")
axes[0].set_ylabel("Shear Force (kN)")
axes[0].grid(True)

# Plot Bending Moment Diagram
axes[1].plot(x, bending_moment, color='r', marker='s', linestyle='-')
axes[1].axhline(0, color='black', linewidth=0.8)
axes[1].set_title("Bending Moment Diagram (BMD)")
axes[1].set_xlabel("Beam Length (m)")
axes[1].set_ylabel("Bending Moment (kNm)")
axes[1].grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

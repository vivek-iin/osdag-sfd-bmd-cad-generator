# Task -1 Developed a Python program using PyPlot to generate Bending Moment and Shear Force Diagrams based on values from an Excel sheet.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SFS_Screening_SFDBMD.xlsx.csv")
```

```python
# Extract relevant columns
positions = df["Distance (m)"]
shear_force = df["SF (kN)"]
bending_moment = df["BM (kN-m)"]
```

```python
# Create the plots
plt.figure(figsize=(12, 5))

# Shear Force Diagram (SFD)
plt.subplot(1, 2, 1)
plt.plot(positions, shear_force, marker='o', linestyle='-', color='b', label="Shear Force")
plt.axhline(0, color='black', linewidth=1)
plt.xlabel("Position along Beam (m)")
plt.ylabel("Shear Force (kN)")
plt.title("Shear Force Diagram (SFD)")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
# Bending Moment Diagram (BMD)
plt.subplot(1, 2, 2)
plt.plot(positions, bending_moment, marker='o', linestyle='-', color='r', label="Bending Moment")
plt.axhline(0, color='black', linewidth=1)
plt.xlabel("Position along Beam (m)")
plt.ylabel("Bending Moment (kN-m)")
plt.title("Bending Moment Diagram (BMD)")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
